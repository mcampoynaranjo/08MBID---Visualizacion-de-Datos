```python
from windrose import WindroseAxes
from matplotlib import pyplot as plt
import matplotlib.cm as cm
import numpy as np
import pandas as pd
```

```python
#Carga del dataset

df = pd.read_excel('DS wind.xlsx')
```

```python
wd, ws = df.direccion, df.velocidad
```

```python
#Representacion
ax = WindroseAxes.from_ax()
ax.contourf(wd, ws, cmap=cm.cool, lw=1, normed=True, nsector=9, bins=10)
ax.set_legend(loc='best', units = 'km/h', fontsize = 'medium')
plt.savefig("Rosa de vientos.jpg", dpi = 600)
```
